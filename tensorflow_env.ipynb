# Check Install for Tensorflow Configuration

**This notebook checks tensorflow, matplotlib, numpy, pillow, and pandas to be working properly.**

**Check which version of TF is installed**

**Check if numpy and plotting is working.**

```
#%matplotlib inline
# import needed to check version, inline used normally
import matplotlib
import pylab
import numpy as np

print("You have matplotlib version %s" % matplotlib.__version__)
print("You have np version %s" % np.__version__)

# create some data using numpy. y = x * 0.1 + 0.3 + noise
x = np.random.rand(100).astype(np.float32)
noise = np.random.normal(scale=0.01, size=len(x))
y = x * 0.1 + 0.3 + noise

# plot it
pylab.plot(x, y, '.')
```

**Check is pillow is working ok.**

```
import PIL
# just need PIL.image, but this is for testing
import numpy as np
from matplotlib.pyplot import imshow

print("You have pillow version %s" % PIL.__version__)

image_array = np.random.rand(200,200,3) * 255
img = PIL.Image.fromarray(image_array.astype('uint8')).convert('RGBA')
imshow(np.asarray(img))
```
